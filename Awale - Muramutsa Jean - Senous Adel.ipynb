```python
def Awale():
    plateau=[[4,4,4,4,4,4],[4,4,4,4,4,4]]
    recolte=[0,0]
    nouvellePartie=1
    
    while nouvellePartie==1:
        joueur=randint(0,1) #le premier à jouer est choisi au hasard
        autre=switchJoueur(joueur)
        print("Le joueur",joueur+1,"commence en premier.")
        
        affiche(plateau)
        
        nbTours=0
        totalGraines= sum(plateau[joueur])+sum(plateau[autre])
        joueurInit=joueur
        disqual=0
        
        
        while recolteMoitie(recolte)==0 and totalGraines>5 and sum(plateau[joueur])!=0 and disqual==0:
            nbTours+=1
            print("Tour",nbTours)
            print("Joueur",joueur+1)
            
            choix=choixTrou(plateau,joueur)
            joueurInit=joueur #joueurInit conserve la valeur de "joueur" car celle-ci peut changer après l'utilisation de la fonction "tour"
    
            
            if choix>=0: #choix renvoie -1 si le joueur est disqualifié
                tour(plateau,recolte,joueur,autre,choix)
                joueur=switchJoueur(joueurInit) 
                autre=switchJoueur(joueur)
                totalGraines=sum(plateau[joueur])+sum(plateau[autre])
            else:
                disqual+=1
        
        finPartie(plateau,recolte,joueur,autre,totalGraines,disqual)
        
        print("Voulez-vous recommencer la partie? (y/n)")
        encore=input()
        if encore=='y':
            recolte=[0,0]
            plateau=[[4,4,4,4,4,4],[4,4,4,4,4,4]]
        if encore=='n':
            nouvellePartie=0
            print("A bientot!")
```

```python
def tour(plateau,recolte,joueur,autre,choix):
    joueurInit=joueur
    trou=choix #le trou d'où le joueur prend les graines avant de les semer
    recolteTour=0
    graines=plateau[joueur][choix]
    plateau[joueur][choix]=0 #le trou sélectionné par le joueur est vidé
    i=1 #compteur de boucle
    while graines-i>=0:
        trou=(choix+i)%6
        if trou==0: #le joueur sème dans le camp de l'autre
            joueur=switchJoueur(joueur)
        if i%12==0:#tour complet
            graines+=1   
        else:
            plateau[joueur][trou]+=1
        
        i+=1
    
    if joueur==autre: #la dernière graine a été semée dans le camp de l'autre joueur
        while trou>=0 and (plateau[joueur][trou]==2 or plateau[joueur][trou]==3):
            recolteTour+=plateau[joueur][trou]
            plateau[joueur][trou]=0
            trou-=1
        
        recolte[joueurInit]+=recolteTour
    
    affiche(plateau)
    print("Recolte joueur 1:",recolte[0])
    print("Recolte joueur 2:",recolte[1])
```

```python
def choixTrou(plateau,joueur):
    print("Choisis un trou (chiffre entre 1 et 6)")
    trou = int(input())
    erreur=0
    while erreur<2 and (trou<1 or trou>6 or plateau[joueur][trou-1]<=0):
        erreur+=1
        print("Erreur")
        if trou<1 or trou>6:
            print("Le nombre choisi n'est pas compris entre 1 et 6")
        if trou>=1 and trou<=6 and plateau[joueur][trou-1]<=0:
            print("Le trou choisi est vide")
        if erreur<2: #une seule autre chance
            print("Nouveau choix:")
            trou=int(input())
    
    if erreur>1: #disqualification
        trou=0
        
    trou-=1 
    
    return trou
```

```python
def finPartie(plateau,recolte,joueur,autre,totalGraines,disqual):
    
    #cas un joueur a la moitié des graines
    gagnant = recolteMoitie(recolte) #la fonction retourne 1 si le joueur 1 a gagné ou 2 si le joueur 2 a gagné
    if gagnant!=0:
        print("Fin de partie.")
        print("Victoire joueur ",gagnant," : a récolté plus de la moitié des graines (",recolte[gagnant-1]," graines).")
        print("Récolte finale du joueur 1: ",recolte[0])
        print("Récolte finale du joueur 2: ",recolte[1])
    
    #cas pas assez de graines
    if totalGraines<=5:
        print("Fin de partie. Il ne reste plus assez de graines sur le plateau.")
        if recolte[0]>recolte[1]:
            print("Victoire joueur 1 (",recolte[0]," graines).")
        if recolte[0]<recolte[1]:
            print("Victoire joueur 2 (",recolte[1]," graines).")
        print("Récolte finale du joueur 1: ",recolte[0])
        print("Récolte finale du joueur 2: ",recolte[1])
        
    #cas un joueur n'a plus de graines
    if sum(plateau[joueur])==0:
        recolte[autre]+=sum(plateau[autre])
        print("Fin de partie. Le joueur ",joueur+1," n'a plus de graines dans son camp.")
        print("Victoire joueur ",autre+1," (",recolte[autre]," graines)")
    
    #cas disqualification
    if disqual!=0:
        print("Fin de partie. Joueur ",joueur+1," disqualifié.")
        print("Victoire joueur ",autre+1,".")
```

```python
def switchJoueur(joueur): #joueur vaut 0 ou 1
    if joueur==0:
        joueur=1
    else:
        joueur=0
    return joueur
```

```python
def recolteMoitie(recolte):
    if recolte[0]>=25:
        return 1 #le joueur 1 a gnagné
    if recolte[1]>=25:
        return 2 #Le joueur 2 a gagné
    else:
        return 0 
```

```python
def affiche(plateau):
    print("Camp joueur 1:",plateau[0])
    print("Camp joueur 2:",plateau[1])
```
